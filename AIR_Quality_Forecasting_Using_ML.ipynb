## Preprocess new data

### Subtask:
Apply the same preprocessing steps to the new data as were applied to the original data. This includes handling missing values and converting 'Date' and 'Time' to the correct format, and combining them into a 'ds' column.


**Reasoning**:
Applying the preprocessing steps to the new data as described in the instructions.



```python
# 1. Remove the last two columns from the new_air_quality_data DataFrame.
new_air_quality_data = new_air_quality_data.iloc[:, :-2]

# 2. Replace all occurrences of -200 with NaN in the new_air_quality_data DataFrame.
new_air_quality_data = new_air_quality_data.replace(to_replace=-200, value=np.nan)

# 3. Fill missing values in the numeric columns of new_air_quality_data with the mean of their respective columns.
numeric_cols_new = new_air_quality_data.select_dtypes(include=np.number).columns
new_air_quality_data[numeric_cols_new] = new_air_quality_data[numeric_cols_new].fillna(new_air_quality_data[numeric_cols_new].mean())

# 4. Drop rows with missing values in the 'Date' or 'Time' columns of new_air_quality_data.
new_air_quality_data.dropna(subset=['Date', 'Time'], inplace=True)

# 5. Convert the 'Date' column in new_air_quality_data to datetime objects.
new_air_quality_data['Date'] = pd.to_datetime(new_air_quality_data['Date'])

# 6. Replace '.' with ':' in the 'Time' column of new_air_quality_data.
new_air_quality_data['Time'] = new_air_quality_data['Time'].str.replace('.', ':')

# 7. Create a new column named 'ds' in new_air_quality_data by concatenating the 'Date' and 'Time' columns, separated by a space.
new_air_quality_data['ds'] = new_air_quality_data['Date'].astype(str) + ' ' + new_air_quality_data['Time'].astype(str)

new_air_quality_data.head()
```
